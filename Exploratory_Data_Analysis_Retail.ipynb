**In both the aboce cases we can consider the values that are out of range as Outliers so will be inspecting them and accordingly take dicision.**

```python
# Products having sales greater than 5000 

len(superstore[superstore.Sales > 5000])
```

**As we can see there are 19 rows having Sales greater than 5000 they will be considered as outliers and also as the count of outliers is very low we will be dropping them.**

```python
# Drop the rows having Sales greater than 5000

superstore = superstore[~(superstore["Sales"] > 5000)]
```

```python
# Re-plot the Sales box plot for inspection

sns.boxplot(superstore["Sales"])
plt.title("Sales Analysis\n", fontdict={"fontsize":15, "fontweight":6, "color":"darkgreen"})
plt.xlabel("\nSales", fontdict={"fontsize":10, "fontweight":5, "color":"red"})
plt.show()
```

**We also saw that the profit for most of the products in dataset ranges between 0 to 2000 and loss ranges between 0 to -2000**

```python
# Inspecting the outliers in Profit column

len(superstore[(superstore["Profit"] < -2000) | (superstore["Profit"] > 2000)])
```

**As we can see that count of outliers is very low we will be dropping those rows having Profit less than -2000 and greater than 2000**

```python
# Drop the rows having Profit less than -2000 and greater than 2000

superstore = superstore[ (superstore["Profit"] > -2000) & (superstore["Profit"] < 2000)]
```

```python
# Inspecting the Discount column value counts

superstore.Discount.value_counts()
```

**We will be converting the discount column into categorical column for better insights.**

```python
# Converting the Discount column into categorical column for better insights

#Create a function that returns Low Discount when Discount is less than 0.2, returns Moderate Discount when less than 0.5 and returns High Discount when other

def discount_func(x):
    if x <= 0.2:
        return "Low Discount"
    elif x <= 0.5:
        return "Moderate Discount"
    else:
        return "High Discount"

superstore["Discount category"] = superstore.Discount.apply(discount_func)
```

```python
superstore.head()
```

```python
# Plotting Matplotlib scatter plot for Discount category vs Profit

plt.figure(figsize=[6,5])
plt.scatter(superstore["Discount category"], superstore.Profit, alpha=0.4, s=100, color="purple")
plt.title("Discount vs Profit\n", fontdict={"fontsize":15, "fontweight":6, "color":"darkgreen"})
plt.xlabel("\nDiscount", fontdict={"fontsize":10, "fontweight":5, "color":"red"})
plt.ylabel("Profit\n", fontdict={"fontsize":10, "fontweight":5, "color":"red"})
plt.show()
```

***Inference:***
From the above scatter plot we can see that:

* Products having low discount (i.e. less than 20%) are more likely to make profit. 
* Products having moderate and high discount (i.e. greater than 20%) have made less profit and more loss. 

### Multi-variate Analysis

**We will be creating a new column say Target column which will be 1 if the product has made profit and 0 for loss.**

```python
# Create new column Target column which is 1 if Product is profitable and 0 if Loss

superstore["Profit_or_Loss"] = superstore["Profit"].apply(lambda x : 1 if x > 0 else 0)
superstore.head()
```

```python
# Separate the Profit and Loss into different dataframes

superstore_1 = superstore[superstore["Profit_or_Loss"] == 1]
superstore_0 = superstore[superstore["Profit_or_Loss"] == 0]
```

```python
# creating a list of categorical column names

categoric_col = list(set(superstore.columns) - set(superstore.describe()))
categoric_col 
```

```python
# As value count of column City and State is too high we will not be inspoecting these columns through bar plots

categoric_col.remove("City")
categoric_col.remove("State")
```

**The below Function "plotting" will return below three plots:**

*1. Subplot : Pie chart for categorical column in terms of percentage.*

*2. Subplot : Bar chart for the value counts of each category in that categorical column with respect to profit and loss.*

*3. Subplot : Bar chart for percentage of each category with respect to profit and loss.*

```python
#Creating a function to plot three subplots mentioned above

def plotting(train, train0, train1, column):
    
    train = train
    train_0 = train0
    train_1 = train1
    col = column
    
    fig = plt.figure(figsize=(10,10))
    
    ax1 = plt.subplot(221)
    train[col].value_counts().plot.pie(autopct = "%1.0f%%", ax=ax1)
    #plt.title('Plotting data for the column: '+ column)
    plt.title("Percentage of each category in "+ column, fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    

    
    ax2 = plt.subplot(222)
    sns.countplot(x= column, hue = "Profit_or_Loss", data = train, ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Plotting data for Profit_or_Loss in terms of total count', fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    plt.xlabel(column, fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    plt.ylabel("Counts", fontdict={"fontsize":10, "fontweight":5, "color":"brown"})


    ax3 = plt.subplot(223)
    df = pd.DataFrame()
    df['0']= ((train_0[col].value_counts())/len(train_0))
    df['1']= ((train_1[col].value_counts())/len(train_1))
    df.plot.bar(ax=ax3)
    plt.title('Plotting data for target in terms of percentage', fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    plt.xlabel(column, fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    plt.ylabel("Percentage", fontdict={"fontsize":10, "fontweight":5, "color":"brown"})



    fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

    plt.show()
```

```python
# Calling the above function to plot three different plots as explained above

no = 1
for col in categoric_col:
    print(no, col)
    plotting(superstore, superstore_1, superstore_0, col)
    print("\n=============================================================================================\n")
    no+=1
```

***Inferences :***

* West region has high product demand whereas South region has lowest product demand. 
* We can see Central region has highest profit percentage. East region has percentage of profit slightly higher than Loss. West and South regions percentage of loss is more than profit.

* Most of the products have low discount.
* Compared to loss, count of profit in Low discount is high but percentage of profit of profit is high in Moderate discount.

* Count of Office Supplies is higher than Furniture and Technology.
* Category Furniture has high percentage of Profit whereas Office Supplies and Technology have loss percentage higher than profit.

* Count of Standard class ship mode is highest.
* Also profit percentage is higher than that of Loss in Standard class and alomost same in First class.

* Profit percentage is higher compared to Loss in Sub-categories - Binders, Storage, Chairs, Bookcases, Tables and Machines.

* Higest count in Segment is for Consumers.
* Also the profit percentage of Consumer segment is higher than Loss.

**Now we will be inspecting the profit by Catgory and Sub-category groups**

```python
# Inspecting the Profit for Category and sub-categories 

product_wise = superstore.groupby(by = ["Category","Sub-Category"]).sum()
```

```python
# Profitable category and sub-category

product_wise.sort_values(by="Profit", ascending = False)
```

***In the above table shows Profit for a given quantity of that Category and Sub-category pair. But we should find the Profit per quantity so that the Category and Sub-category pair will be at a comparable scale.***

```python
# Creating column Profit_per_product by dividing Profit by Quantity

product_wise["Profit_per_product"] = product_wise["Profit"] / product_wise["Quantity"]
```

**Top profitable products**

```python
# Top profitable products

product_wise.sort_values("Profit_per_product", ascending = False).head(10)
```

***Inferences :***

Listed above are the top 10 Profitable products.
